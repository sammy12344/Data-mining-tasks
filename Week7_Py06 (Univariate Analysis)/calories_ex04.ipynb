```python
import pandas as pd

df = pd.read_csv('clean_data.csv')
df.head()
```

```python
df.rename( columns={'Unnamed: 0':'Date'}, inplace=True )
df.head()
```

```python
# Adding numerical data
q1 = df['Calories'].quantile(q=0.25)
q2 = df['Calories'].median()
q3 = df['Calories'].quantile(q=0.75)
q4 = df['Calories'].max()
```

```python
df.describe()
```

```python
df['Category'] = pd.cut(x=df['Calories'], bins=[0, q1, q2, q3, q4], labels=['very few', 'few', 'normal', 'high'])
df.head()
```

```python
# Grouping by mean
df.groupby(['Category']).mean()
```
